```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_pato = cv2.imread('pato.jpg', cv2.IMREAD_GRAYSCALE)
img_puntos = cv2.imread('puntos.png', cv2.IMREAD_GRAYSCALE)

img_color_pato = cv2.imread('pato.jpg')
img_gris_pato = cv2.cvtColor(img_color_pato, cv2.COLOR_BGR2GRAY)

img_color_puntos = cv2.imread("puntos.png")
img_gris_puntos = cv2.cvtColor(img_color_puntos, cv2.COLOR_BGR2GRAY)

img_puntos_redimensionada = cv2.resize(img_gris_puntos, (img_gris_pato.shape[1], img_gris_pato.shape[0]))

img_combinada = cv2.addWeighted(img_gris_pato, 1, img_puntos_redimensionada, 1, 0)

plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title('pato')
plt.imshow(img_pato, cmap='gray')

plt.subplot(1, 3, 2)
plt.title('puntos')
plt.imshow(img_puntos, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('combinada')
plt.imshow(img_combinada, cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import skimage.measure as ms

def red_counter(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)
    labels = ms.label(binary, connectivity=2)
    count = np.max(labels)
    return count, binary

def orange_counter(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 220, 255, cv2.THRESH_BINARY_INV)
    labels = ms.label(binary, connectivity=2)
    count = np.max(labels)
    return count, binary

def yellow_counter(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 230, 255, cv2.THRESH_BINARY_INV)
    labels = ms.label(binary, connectivity=2)
    count = np.max(labels)
    return count, binary

def green_counter(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 230, 232, cv2.THRESH_BINARY_INV)
    labels = ms.label(binary, connectivity=2)
    count = np.max(labels)
    return count, binary

def blue_counter(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, binary = cv2.threshold(gray, 230, 245, cv2.THRESH_BINARY_INV)
    labels = ms.label(binary, connectivity=2)
    count = np.max(labels)
    return count, binary

img = cv2.imread('colores.jpg')

red_zone = img[1:1000, 1:300, :] 
orange_zone = img[1:1000, 300:600, :] 
yellow_zone = img[1:1000, 600:800, :] 
green_zone = img[1:1000, 850:1070, :] 
blue_zone = img[1:1000, 1100:1400, :] 

red_count, red_bin = red_counter(red_zone)
orange_count, orange_bin = orange_counter(orange_zone)
yellow_count, yellow_bin = yellow_counter(yellow_zone)
green_count, green_bin = green_counter(green_zone)
blue_count, blue_bin = blue_counter(blue_zone)

plt.figure(figsize=(10, 5))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(2, 3, 2)
plt.title(f'Objetos: {red_count}')
plt.imshow(cv2.cvtColor(red_zone, cv2.COLOR_BGR2RGB))

plt.subplot(2, 3, 3)
plt.title(f'Objetos: {orange_count}')
plt.imshow(cv2.cvtColor(orange_zone, cv2.COLOR_BGR2RGB))

plt.subplot(2, 3, 4)
plt.title(f'Objetos: {yellow_count}')
plt.imshow(cv2.cvtColor(yellow_zone, cv2.COLOR_BGR2RGB))

plt.subplot(2, 3, 5)
plt.title(f'Objetos: {green_count}')
plt.imshow(cv2.cvtColor(green_zone, cv2.COLOR_BGR2RGB))

plt.subplot(2, 3, 6)
plt.title(f'Objetos: {blue_count}')
plt.imshow(cv2.cvtColor(blue_zone, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

imagen = cv2.imread('girl.jpg')

imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

imagen_borrosa = cv2.GaussianBlur(imagen_gris, (5, 5), 1.4)

bordes = cv2.Canny(imagen_borrosa, 20, 80)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('imagen original')

plt.subplot(1, 2, 2)
plt.imshow(bordes, cmap='gray')
plt.title('imagen con canny')

plt.show()
```
