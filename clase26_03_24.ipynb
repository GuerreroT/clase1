```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('paisaje.jpg', cv2.IMREAD_GRAYSCALE)

histograma, _ = np.histogram(img, 256, [0, 256])
funcion_dist = histograma.cumsum()

hst_normalizado = funcion_dist * histograma.max() / funcion_dist.max()

img_ecualizada = cv2.equalizeHist(img)

histograma_ecualizado, _ = np.histogram(img_ecualizada, 256, [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.plot(hst_normalizado, color='black')
plt.title('Histograma de la Imagen Original')

plt.subplot(2, 2, 3)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')

plt.subplot(2, 2, 4)
plt.plot(histograma_ecualizado, color='black')
plt.title('Histograma de la Imagen Ecualizada')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la imagen original y aplicar CLAHE
img_original = cv2.imread('paisaje.jpg')
img_ecualizada = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8)).apply(cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY))

# Calcular los histogramas de color de la imagen original
hist_original_b = cv2.calcHist([img_original], [0], None, [256], [0, 256])
hist_original_g = cv2.calcHist([img_original], [1], None, [256], [0, 256])
hist_original_r = cv2.calcHist([img_original], [2], None, [256], [0, 256])

# Calcular el histograma de la imagen ecualizada
hist_ecualizada = cv2.calcHist([img_ecualizada], [0], None, [256], [0, 256])

# Mostrar las imgenes y sus histogramas
plt.figure(figsize=(15, 10))

# Imagen Original
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.plot(hist_original_b, color='blue')
plt.plot(hist_original_g, color='green')
plt.plot(hist_original_r, color='red')
plt.title('Histograma de la Imagen Original')

# Imagen Ecualizada
plt.subplot(2, 2, 3)
plt.imshow(img_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')

plt.subplot(2, 2, 4)
plt.plot(hist_ecualizada, color='black')
plt.title('Histograma de la Imagen Ecualizada')

plt.tight_layout()
plt.show()

```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('paisaje.jpg')

histograma, _ = np.histogram(img, 256, [0, 256])
funcion_dist = histograma.cumsum()

hst_normalizado = funcion_dist * histograma.max() / funcion_dist.max()

img_ecualizada = cv2.equalizeHist(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))

histograma_ecualizado, _ = np.histogram(img_ecualizada, 256, [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.plot(hst_normalizado, color='red')
plt.title('Histograma de la Imagen Original')

plt.subplot(2, 2, 3)
plt.imshow(img_ecualizada)
plt.title('Imagen Ecualizada'  )

plt.subplot(2, 2, 4)
plt.plot(histograma_ecualizado, color='blue')
plt.title('Histograma de la Imagen Ecualizada')

plt.tight_layout()
plt.show()
```
