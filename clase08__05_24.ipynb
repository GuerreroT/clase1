opencv

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

def etiquetar(image):
    _, image_bin = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
    contorno, _ = cv2.findContours(image_bin, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    img_et = np.zeros_like(image, dtype=np.uint8)
    for i, contorno1 in enumerate(contorno):
        cv2.drawContours(img_et, [contorno1], -1, (i+1), -1)
    return img_et

img = cv2.imread('paisaje.jpg', cv2.IMREAD_GRAYSCALE)

image_fin = etiquetar(img)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('imagen original')

plt.subplot(1, 2, 2)
plt.imshow(image_fin, cmap='jet')  
plt.title('imagen etiquetada')

plt.show()
```

scikit-image

```python
import cv2
import skimage.measure
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('girl.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar umbralización
_, binaria = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

img_etiquetada = skimage.measure.label(binaria, connectivity=1)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('imagen original')

plt.subplot(1, 2, 2)
plt.imshow(img_etiquetada, cmap='jet')  
plt.title('imagen etiquetada')

plt.show()
```
