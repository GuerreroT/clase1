dilatacion con binaria

```python
import cv2 
import matplotlib.pyplot as plt 

img = cv2.imread('girl.jpg', cv2.IMREAD_GRAYSCALE)
_, img_bin=cv2.threshold(img, 127,255,cv2.THRESH_BINARY)
mascara=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5) )
img_fin=cv2.dilate(img_bin, mascara)

plt.subplot(121)
plt.imshow(img_bin, cmap='gray')
plt.subplot(122)
plt.imshow(img_fin, cmap='gray')

plt.show()
```

reapertura de cerradura

```python
import cv2 
import matplotlib.pyplot as plt 

img = cv2.imread('cancer.jpg', cv2.IMREAD_GRAYSCALE)

mascara=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(9,9) )

cerradura=cv2.morphologyEx(img,cv2.MORPH_CLOSE,mascara)


plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.subplot(122)
plt.imshow(cerradura, cmap='gray')

plt.show()
```

reapertura de apertura

```python
import cv2 
import matplotlib.pyplot as plt 

img = cv2.imread('ruido.jpg', cv2.IMREAD_GRAYSCALE)

mascara=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5) )

apertura=cv2.morphologyEx(img,cv2.MORPH_OPEN,mascara)

k_reconstruccion=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

img_fin =cv2.morphologyEx(apertura,cv2.MORPH_CLOSE, k_reconstruccion)
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.subplot(122)
plt.imshow(img_fin, cmap='gray')
plt.title('reconstruccion open')

plt.show()
```

cerradura


```python
import cv2 
import matplotlib.pyplot as plt 

img = cv2.imread('ruido.jpg', cv2.IMREAD_GRAYSCALE)

mascara=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5) )

cerradura=cv2.morphologyEx(img,cv2.MORPH_CLOSE,mascara)

k_reconstruccion=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))

img_fin =cv2.morphologyEx(cerradura,cv2.MORPH_OPEN, k_reconstruccion)
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('original')
plt.subplot(122)
plt.imshow(img_fin, cmap='gray')
plt.title('reconstruccion Close')

plt.show()
```

operacion mophologia de cerradura

```python
import cv2 
import matplotlib.pyplot as plt 
from skimage.morphology import closing, opening, disk

img = cv2.imread('ruido.jpg', cv2.IMREAD_GRAYSCALE)

mascara = disk(2)
k_reconstruccion = disk(1)

cerradura = closing(img, mascara)
img_fin = opening(cerradura, k_reconstruccion)

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(122)
plt.imshow(img_fin, cmap='gray')
plt.title('Reconstruccion Close')

plt.show()
```
