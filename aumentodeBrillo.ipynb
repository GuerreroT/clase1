Investiga los algoritmos de aumento de brillo de una imagen y apl√≠calos en una imagen haciendo uso de Python y jupyter 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
def aumentar_luminosidad(imagen, aumento):
    # Convertir la imagen a escala de grises
    imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    # Aumentar la luminosidad multiplicando cada p√≠xel por el aumento
    nueva_imagen = cv2.addWeighted(imagen_gris, 1, np.zeros(imagen_gris.shape, imagen_gris.dtype), 0, aumento)
    return nueva_imagen

# Cargar la imagen
imagen = cv2.imread('girl.jpg')

# Aumentar la luminosidad
imagen_aumentada = aumentar_luminosidad(imagen, aumento=100)
imagen_reducida = aumentar_luminosidad(imagen, aumento=-100)

# Mostrar las imagenes
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(imagen_aumentada, cmap='gray')
plt.title('Aumento de luminosidad')

plt.subplot(1, 3, 3)
plt.imshow(imagen_reducida, cmap='gray')
plt.title('reducir de luminosidad')

plt.show()
```
