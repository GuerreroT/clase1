Ejercicio 2.
Genera un degradado lineal a colores.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Generar una imagen con un gradiente lineal
altura = 200
ancho = 200
imagen_gradiente = np.zeros((altura, ancho), dtype=np.uint8)
for y in range(altura):
    for x in range(ancho):
        imagen_gradiente[y, x] = int(255 * y / ancho)

plt.figure(figsize=(15,10))

plt.subplot(1, 1, 1)
plt.imshow(imagen_gradiente)
plt.title("imagen_gradiente")

plt.show()
```

Ejercicio 3.
Genera una imagen semejante a un tablero de ajedrez tomando en cuenta que cada casilla debe medir 100x100 px 

```python
import numpy as np
import matplotlib.pyplot as plt

# Dimensiones del tablero de ajedrez (8x8)
filas = 8
columnas = 8
tamano_casilla = 100  # Tama√±o de cada casilla en p√≠xeles

# Crear una matriz para representar el tablero de ajedrez
tablero = np.zeros((filas * tamano_casilla, columnas * tamano_casilla), dtype=np.uint8)

# Alternar entre blanco y negro para las casillas
for fila in range(filas):
    for columna in range(columnas):
        if (fila + columna) % 2 == 0:
            tablero[fila * tamano_casilla: (fila + 1) * tamano_casilla,
                    columna * tamano_casilla: (columna + 1) * tamano_casilla] = 255  # Blanco
        else:
            tablero[fila * tamano_casilla: (fila + 1) * tamano_casilla,
                    columna * tamano_casilla: (columna + 1) * tamano_casilla] = 0  # Negro

# Mostrar la imagen del tablero de ajedrez
plt.figure(figsize=(8, 8))
plt.imshow(tablero, cmap='gray')  # Usar colormap 'gray' para mostrar en blanco y negro
plt.title('Tablero de Ajedrez')
plt.show()
```

Ejercicio 4.
Tomando en cuenta el ejercicio 2 aplica la ecualizaci√≥n del histograma y muestrael mismo viendo las diferencias entre im√°genese histogramas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen
imagen_original = cv2.imread('imagen2.jpg', cv2.IMREAD_GRAYSCALE)

# Ecualizar el histograma de la imagen
imagen_ecualizada = cv2.equalizeHist(imagen_original)

# Calcular los histogramas de las im√°genes original y ecualizada
hist_original = cv2.calcHist([imagen_original], [0], None, [256], [0, 256])
hist_ecualizada = cv2.calcHist([imagen_ecualizada], [0], None, [256], [0, 256])

# Mostrar la imagen original y la imagen ecualizada junto con sus histogramas
plt.figure(figsize=(10, 8))

# Imagen original
plt.subplot(2, 2, 1)
plt.imshow(imagen_original, cmap='gray')
plt.title('Imagen Original')

plt.subplot(2, 2, 2)
plt.plot(hist_original, color='black')
plt.title('Histograma Original')
plt.xlabel('Intensidad de p√≠xeles')
plt.ylabel('Frecuencia')

# Imagen ecualizada
plt.subplot(2, 2, 3)
plt.imshow(imagen_ecualizada, cmap='gray')
plt.title('Imagen Ecualizada')

plt.subplot(2, 2, 4)
plt.plot(hist_ecualizada, color='black')
plt.title('Histograma Ecualizado')
plt.xlabel('Intensidad de p√≠xeles')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

Ejercicio 5.
Tomando en cuenta la imagen1.jpg aplica un degradado lineal en ella.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen original
imagen_original = cv2.imread('imagen1.jpg')

# Obtener dimensiones de la imagen
alto, ancho, _ = imagen_original.shape

# Generar un gradiente lineal
gradiente = np.zeros((alto, ancho), dtype=np.uint8)
for y in range(alto):
    for x in range(ancho):
        gradiente[y, x] = int(255 * x / ancho)  # Valor de p√≠xel basado en la posici√≥n horizontal

# Combinar el gradiente con la imagen original
imagen_con_degradado = cv2.merge([imagen_original[:,:,0] + gradiente, 
                                  imagen_original[:,:,1] + gradiente, 
                                  imagen_original[:,:,2] + gradiente])

# Mostrar la imagen original y la imagen con degradado utilizando Matplotlib
plt.figure(figsize=(10, 5))

# Mostrar la imagen original
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')

# Mostrar la imagen con degradado
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(imagen_con_degradado, cv2.COLOR_BGR2GRAY))
plt.title('Imagen con Degradado')

plt.tight_layout()
plt.show()

```
