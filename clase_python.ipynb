```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

image =cv2.imread('cat.jpg',cv2.IMREAD_GRAYSCALE)
#guandamos profundidad original para mostrar
original_img=image.dtype
print('profundidad original de color : ',original_img)
#transformacion de imagen a 16bits
image_16bits=np.uint16(image)
nueva_img =image_16bits.dtype

print('profundidad nueva de color : ',nueva_img)

#guardamos la imagen tratada
cv2.imwrite('image16bits.jpg', image_16bits)

img_guardada=cv2.imread('image16bits.jpg',cv2.IMREAD_UNCHANGED)

plt.figure(figsize =(10,5))

plt.subplot(121)
plt.title('original')
plt.imshow(image,cmap='gray')

plt.subplot(122)
plt.title('imagen tratada 16bits')
plt.imshow(img_guardada,cmap='gray')

plt.show()
```
