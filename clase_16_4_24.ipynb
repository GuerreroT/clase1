```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('cancer.jpg', 0)

mascara1 = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
mascara2 = np.array([[1, 1, 1], [1, -7, 1], [1, 1, 1]])
mascara3 = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

imgfin1 = cv2.filter2D(img, -1, mascara1)
imgfin2 = cv2.filter2D(img, -1, mascara2)
imgfin3 = cv2.filter2D(img, -1, mascara3)

plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(img, cmap='gray')
plt.title('Imagen Original')

plt.subplot(132)
plt.imshow(imgfin1, cmap='gray')
plt.title('Filtro 1')

plt.subplot(133)
plt.imshow(imgfin2, cmap='gray')
plt.title('Filtro 2')

plt.show()

_, img_bin = cv2.threshold(imgfin3, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.imshow(imgfin3, cmap='gray')
plt.title('Filtro 3')

plt.subplot(122)
plt.imshow(img_bin, cmap='gray')
plt.title('Imagen Binarizada')

plt.show()
```
