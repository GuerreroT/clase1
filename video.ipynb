```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la imagen
imagen = cv2.imread('girl2.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar el operador Laplaciano
laplaciano = cv2.Laplacian(imagen, cv2.CV_64F)

# Convertir los valores de pxeles a valores absolutos
laplaciano = np.uint8(np.absolute(laplaciano))

# Mostrar la imagen original y la imagen con bordes detectados
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1), plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 2, 2), plt.imshow(laplaciano, cmap='gray')
plt.title('Bordes Detectados con Laplaciano'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Leer la imagen en escala de grises
imagen = cv2.imread('girl2.jpg', cv2.IMREAD_GRAYSCALE)

# Aplicar filtro Gaussiano
imagen_suavizada = cv2.GaussianBlur(imagen, (3, 3), 0)  # Puedes ajustar el tamao del kernel (3, 3) y la desviacin estndar (0)

# Aplicar el operador Laplaciano
laplaciano = cv2.Laplacian(imagen_suavizada, cv2.CV_64F)

# Convertir los valores de pxeles a valores absolutos
laplaciano = np.uint8(np.absolute(laplaciano))

# Mostrar la imagen original y la imagen con bordes detectados
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1), 
plt.imshow(imagen, cmap='gray')
plt.title('Imagen Original')

plt.subplot(1, 2, 2), 
plt.imshow(laplaciano, cmap='gray')
plt.title('Bordes Detectados con LoG')
plt.show()
```
