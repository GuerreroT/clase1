tablero

```python
import numpy as np
import matplotlib.pyplot as plt

filas = 10
columnas = 10
tamano_casilla = 100 

tablero = np.zeros((filas * tamano_casilla, columnas * tamano_casilla), dtype=np.uint8)


for i in range(filas):
    for j in range(columnas):
        if i % 2== 0:
            if j % 2 == 0:
                cl=0
            else:
                cl=127
        else:
            if j % 2 ==0:
                cl=127
            else:
                cl=255
        tablero[i * tamano_casilla:(i + 1) * tamano_casilla, j * tamano_casilla:(j + 1) * tamano_casilla] = cl

plt.imshow(tablero, cmap='gray')

plt.show()
```

triangulo

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('mascara.jpeg')

alto, ancho = img.shape[:2]

mascara = np.zeros_like(img)

vertices = np.array([[ancho//2, 50], [ancho-50, alto-50], [50, alto-50]])
cv2.drawContours(mascara, [vertices], 0, (255, 255, 255), -1)

img_con_mascara = cv2.bitwise_and(img, mascara)

plt.figure(figsize=(12, 10))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img_con_mascara, cv2.COLOR_BGR2RGB))
plt.title('Máscara Triangular')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
num_puntos = 5
tamanos_base = np.random.randint(250, 300, num_puntos)
tamanos = tamanos_base + 200
colores = np.random.rand(255, 3)
coordenadas_x = np.random.rand(num_puntos)
coordenadas_y = np.random.rand(num_puntos)

for i in range(num_puntos):
    plt.scatter(coordenadas_x[i], coordenadas_y[i], s=tamanos[i], c=[colores[i]])

plt.show()
```
