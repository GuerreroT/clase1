deteccion de bordes Roberts 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('girl2.jpg', cv2.IMREAD_GRAYSCALE)
kernel_x = np.array([[1, 0],
                    [0, -1]])

kernel_y = np.array([[0, 1],
                    [-1, 0]])
bordes_x = cv2.filter2D(imagen, -1, kernel_x)
bordes_y = cv2.filter2D(imagen, -1, kernel_y)
bordes = np.sqrt(np.square(bordes_x) + np.square(bordes_y))
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('imagen original')

plt.subplot(2, 2, 2)
plt.imshow(bordes_x, cmap='gray')
plt.title('bordes X ')

plt.subplot(2, 2, 3)
plt.imshow(bordes_y, cmap='gray')
plt.title('bordes Y ')

plt.subplot(2, 2, 4)
plt.imshow(bordes, cmap='gray')
plt.title('bordes Roberts')

plt.show()
```

deteccion de bordes prewitt

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('girl2.jpg', cv2.IMREAD_GRAYSCALE)

kernel_x = np.array([[-1, 0, 1],
                     [-1, 0, 1],
                     [-1, 0, 1]])

kernel_y = np.array([[1, 1, 1],
                     [0, 0, 0],
                     [-1, -1, -1]])

bordes_x = cv2.filter2D(imagen, -1, kernel_x)
bordes_y = cv2.filter2D(imagen, -1, kernel_y)

bordes = np.sqrt(np.square(bordes_x) + np.square(bordes_y))

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('imagen original')

plt.subplot(2, 2, 2)
plt.imshow(bordes_x, cmap='gray')
plt.title('bordes X ')

plt.subplot(2, 2, 3)
plt.imshow(bordes_y, cmap='gray')
plt.title('bordes Y')

plt.subplot(2, 2, 4)
plt.imshow(bordes, cmap='gray')
plt.title('bordes Prewitt')

plt.show()
```

deteccion de bordes sobel

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('girl2.jpg', cv2.IMREAD_GRAYSCALE)

sobel_x = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(imagen, cv2.CV_64F, 0, 1, ksize=3)

bordes = np.sqrt(np.square(sobel_x) + np.square(sobel_y))

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('imagen original')

plt.subplot(2, 2, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('bordes X ')

plt.subplot(2, 2, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('bordes Y ')

plt.subplot(2, 2, 4)
plt.imshow(bordes, cmap='gray')
plt.title('bordes Sobel')

plt.show()
```

deteccion de bordes cany

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('girl2.jpg', cv2.IMREAD_GRAYSCALE)

bordes = cv2.Canny(imagen, 100, 200)  

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(imagen, cmap='gray')
plt.title('imagen original')

plt.subplot(1, 2, 2)
plt.imshow(bordes, cmap='gray')
plt.title('bordes Canny')

plt.show()
```
