```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('michi.jpg', 0)

t1 = np.fft.fft2(img)
t2 = np.fft.fftshift(t1)

fila, columna = img.shape
fila1, columna1 = fila // 2, columna // 2

t2[fila - 5:fila + 5, columna1 - 5:columna1 + 5] = 0
t3 = np.fft.ifftshift(t2)

img_fin = np.fft.ifft2(t3)
img_fin = np.abs(img_fin)

espectro = 20 * np.log(np.abs(t2))

plt.figure(figsize=(12, 10))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(img_fin, cmap='gray')
plt.title('Procesada')

plt.subplot(1, 3, 3)
plt.imshow(espectro, cmap='gray')
plt.title('Espectro')

plt.show()

```
