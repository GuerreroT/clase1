aplicalos con las imagenes recolectadas 

moiré

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread('moire.jpg')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_blurred = cv2.GaussianBlur(img_rgb, (5, 55), 0)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen original')

plt.subplot(1, 2, 2)
plt.imshow(img_blurred)
plt.title('Imagen suavizada')

plt.show()
```

Aliasing 

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread('aliasing.webp')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

sampling_rate = 5  

h, w, _ = img_rgb.shape
resized_img = cv2.resize(img_rgb, (int(w/sampling_rate), int(h/sampling_rate)))

upscaled_img = cv2.resize(resized_img, (w, h), interpolation=cv2.INTER_NEAREST)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(upscaled_img)
plt.title(' Aliasing')

plt.show()
```
