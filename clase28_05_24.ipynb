```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('girl.jpg')
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, mascara=cv2.threshold(gris,127,255,cv2.THRESH_BINARY)

kernel= np.array([[0,1,0],[1,1,1],[0,1,1]],dtype=np.uint8)#personalizable

erosion=cv2.erode(mascara, kernel, iterations=1)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Imagen Original')
plt.imshow(img)

plt.subplot(2, 2, 2)
plt.title('mascara')
plt.imshow(mascara)

plt.subplot(2, 2, 3)
plt.title('erosion')
plt.imshow(erosion)
plt.show()
```

```python
import numpy as np 
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('girl1.jpg')
img_2 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
arr1 = np.array([10, 10, 10])
arr2 = np.array([148, 127, 100])

copia = img_2.copy()
mascara = cv2.inRange(copia, arr1, arr2)
kernel1 = np.ones((3, 3), np.uint8)
erosion1 = cv2.erode(mascara, kernel1, iterations=1)

kernel2 = np.ones((5, 5), np.uint8)
kernel3 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
kernel4 = cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))

erosion2 = cv2.erode(mascara, kernel2, iterations=1)
erosion3 = cv2.erode(mascara, kernel3, iterations=1)
erosion4 = cv2.erode(mascara, kernel4, iterations=1)

plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.title('Imagen Original')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(3, 2, 2)
plt.title('Mascara')
plt.imshow(mascara, cmap='gray')

plt.subplot(3, 2, 3)
plt.title('Erosion con Kernel 3x3')
plt.imshow(erosion1, cmap='gray')

plt.subplot(3, 2, 4)
plt.title('Erosion con Kernel 5x5 Cuadrado')
plt.imshow(erosion2, cmap='gray')

plt.subplot(3, 2, 5)
plt.title('Erosion con Kernel 5x5 Elipse')
plt.imshow(erosion3, cmap='gray')

plt.subplot(3, 2, 6)
plt.title('Erosion con Kernel 5x5 Cruz')
plt.imshow(erosion4, cmap='gray')

plt.tight_layout()
plt.show()
```
