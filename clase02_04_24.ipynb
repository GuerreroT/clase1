Redireccionar imagen

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('girl.jpg')

scale=0.5

img_redim=cv2.resize(img,None,fx=scale,fy=scale)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(img_redim, cmap='gray')
plt.title('imagen redireccionada')

plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('girl1.jpg')

x=60
y=120

matriz_tras=np.float32([[1,0,x],[0,1,y]])

img_tras=cv2.warpAffine(img,matriz_tras,(img.shape[1],img.shape[0]))


plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(matriz_tras, cmap='gray')
plt.title('posicion  matriz')

plt.subplot(1, 3, 3)
plt.imshow(img_tras, cmap='gray')
plt.title('img trasladada')

plt.show()
```

filtro de mascara de figuras 

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('michi.jpg', cv2.IMREAD_GRAYSCALE)

alto, ancho=img.shape[:2]

mascara= np.zeros_like (img)
circulo=(ancho//2,alto//2)

radio=180
cv2.circle(mascara,circulo,radio,(255),-1)
img_fin=cv2.bitwise_and(img,img,mask=mascara)

plt.figure(figsize=(12, 10))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(mascara, cmap='gray')
plt.title('mascara')

plt.subplot(1, 3, 3)
plt.imshow(img_fin, cmap='gray')
plt.title('img con mascara')

plt.show()
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('michi.jpg', cv2.IMREAD_GRAYSCALE)

alto, ancho=img.shape[:2]

mascara= np.zeros_like (img)
circulo=(ancho//2,alto//2)

radio=180

cv2.circle(mascara,circulo,radio,(255),-1)
img_fin=img*(mascara/255)+(img//2)*(1-mascara/255)

plt.figure(figsize=(12, 10))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(1, 3, 2)
plt.imshow(mascara, cmap='gray')
plt.title('mascara')

plt.subplot(1, 3, 3)
plt.imshow(img_fin, cmap='gray')
plt.title('img con mascara')

plt.show()
```
