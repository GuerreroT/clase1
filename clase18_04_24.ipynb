```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def conv_horizontal(img, mascara1):
    mascara=np.ones((1,mascara1),np.float32)/mascara1
    img_conv=cv2.filter2D(img,-1,mascara)
    return img_conv

image=cv2.imread('aliasing.webp')
img_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

img_r=conv_horizontal(img_rgb[:,:,0],111)
img_g=conv_horizontal(img_rgb[:,:,1],111)
img_b=conv_horizontal(img_rgb[:,:,2],111)

img_fin=np.stack([img_r,img_g,img_b],axis=2)

plt.subplot(121)
plt.imshow(img_rgb)

plt.subplot(122)
plt.imshow(img_fin)

plt.show()
```

aplica la convolucion vertical y deagonal en el codigo expresado

vertical

```python
def conv_vertical(img, mascara1):
    mascara = np.ones((mascara1,1), np.float32) / mascara1
    img_conv = cv2.filter2D(img, -1, mascara)
    return img_conv

img_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

img_r_vert=conv_vertical(img_rgb[:,:,0],111)
img_g_vert=conv_vertical(img_rgb[:,:,1],111)
img_b_vert=conv_vertical(img_rgb[:,:,2],111)

img_fin_vert=np.stack([img_r_vert,img_g_vert,img_b_vert],axis=2)

plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Original')

plt.subplot(122)
plt.imshow(img_fin_vert)
plt.title('Vertical')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def conv_diagonal(img, mascara):
    tamano_mascara = len(mascara)
    diago = np.diag(mascara)
    
    img_conv = cv2.filter2D(img, -1, diago)
    return img_conv

image = cv2.imread('aliasing.webp')
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Definir la mascara diagonal
mascara_diagonal = np.ones(111) / 111

img_r = conv_diagonal(img_rgb[:, :, 0], mascara_diagonal)
img_g = conv_diagonal(img_rgb[:, :, 1], mascara_diagonal)
img_b = conv_diagonal(img_rgb[:, :, 2], mascara_diagonal)

img_fin = np.stack([img_r, img_g, img_b], axis=2)

plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Original')

plt.subplot(122)
plt.imshow(img_fin)
plt.title(' Diagonal')

plt.show()
```
